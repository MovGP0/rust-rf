# Media

This is the **rust-rf** edition of the upstream scikit-rf notebook. It demonstrates
creating a distributed transmission-line medium and a physical line. Rust examples use typed values and explicit `Result` handling.

The original notebook contained 2 Python code cells. They have been
replaced by a compact, executable Rust workflow instead of retaining stale Python syntax.
The copied data files and figures remain beside the notebook for extending the example.


## Kernel setup

The first cell adds the local crate and the numerical types used by the example.


```rust
:dep rust-rf = { path = "../../.." }
:dep ndarray = "0.17.2"
:dep num-complex = "0.4.6"
```

## Example


```rust
use rust_rf::media::{DistributedCircuit, LengthUnit, Media};
use rust_rf::{Frequency, FrequencyUnit, SweepType};

let frequency = Frequency::new(
    1.0,
    3.0,
    51,
    FrequencyUnit::GHz,
    SweepType::Linear,
)
.expect("the frequency sweep should be valid");
let medium = DistributedCircuit::from_scalars(
    frequency,
    1.0,
    0.01,
    250.0e-9,
    100.0e-12,
)
.expect("the medium should be valid");
let line = medium
    .line(25.0, LengthUnit::Millimeter)
    .expect("the line should be generated");

println!("{} mm line represented by a {}-port network", 25, line.ports());
```
