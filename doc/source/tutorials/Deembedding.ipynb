# Deembedding

This is the **rust-rf** edition of the upstream scikit-rf notebook. It demonstrates
selecting typed de-embedding algorithms and applying them explicitly. Rust examples use typed values and explicit `Result` handling.

The original notebook contained 2 Python code cells. They have been
replaced by a compact, executable Rust workflow instead of retaining stale Python syntax.
The copied data files and figures remain beside the notebook for extending the example.


## Kernel setup

The first cell adds the local crate and the numerical types used by the example.


```rust
:dep rust-rf = { path = "../../.." }
:dep ndarray = "0.17.2"
:dep num-complex = "0.4.6"
```

## Example


```rust
use rust_rf::calibration::{Deembedding, Open};
use rust_rf::data::DATA;

let open_dummy = DATA.ro_1().expect("the embedded one-port dummy should load");
let measured = DATA.ro_2().expect("the embedded one-port measurement should load");
let algorithm = Open::new(open_dummy, Some("open removal".to_owned()));
let corrected = algorithm
    .deembed(&measured)
    .expect("the open parasitic should be removed");

println!("de-embedded {} frequency points", corrected.frequency_points());
```
