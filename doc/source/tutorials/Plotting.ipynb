# Plotting

This is the **rust-rf** edition of the upstream scikit-rf notebook. It demonstrates
building backend-neutral plot data that can be rendered with the plot feature. Rust examples use typed values and explicit `Result` handling.

The original notebook contained 2 Python code cells. They have been
replaced by a compact, executable Rust workflow instead of retaining stale Python syntax.
The copied data files and figures remain beside the notebook for extending the example.


## Kernel setup

The first cell adds the local crate and the numerical types used by the example.


```rust
:dep rust-rf = { path = "../../.." }
:dep ndarray = "0.17.2"
:dep num-complex = "0.4.6"
```

## Example


```rust
use rust_rf::data::DATA;
use rust_rf::plotting::{Component, Parameter, network_plot};

let network = DATA.ntwk1().expect("the embedded network should load");
let plot = network_plot(
    &network,
    Parameter::Scattering,
    Component::Decibels,
    None,
)
.expect("plot data should be generated");

println!("{} traces; x-axis: {}", plot.series.len(), plot.x_label);
```
