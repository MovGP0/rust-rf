# SOLT

This is the **rust-rf** edition of the upstream scikit-rf notebook. It demonstrates
using the typed calibration trait and explicit run/apply lifecycle. Rust examples use typed values and explicit `Result` handling.

The original notebook contained 2 Python code cells. They have been
replaced by a compact, executable Rust workflow instead of retaining stale Python syntax.
The copied data files and figures remain beside the notebook for extending the example.


## Kernel setup

The first cell adds the local crate and the numerical types used by the example.


```rust
:dep rust-rf = { path = "../../../.." }
:dep ndarray = "0.17.2"
:dep num-complex = "0.4.6"
```

## Example


```rust
use rust_rf::calibration::{Calibration, Normalization};
use rust_rf::data::DATA;

let measured = DATA.ro_1().expect("the embedded one-port network should load");
let mut calibration = Normalization::new(vec![measured.clone(), measured.clone()])
    .expect("the normalization standards should be compatible");
calibration.run().expect("the calibration should run");
let corrected = calibration
    .apply(&measured)
    .expect("the calibration should apply");

println!("corrected {} frequency points", corrected.frequency_points());
```
