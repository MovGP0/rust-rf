# VNA Control

This is the **rust-rf** edition of the upstream scikit-rf notebook. It demonstrates
defining an instrument sweep while keeping live I/O behind the visa feature. Rust examples use typed values and explicit `Result` handling.

The original notebook contained 2 Python code cells. They have been
replaced by a compact, executable Rust workflow instead of retaining stale Python syntax.
The copied data files and figures remain beside the notebook for extending the example.


## Kernel setup

The first cell adds the local crate and the numerical types used by the example.


```rust
:dep rust-rf = { path = "../../../.." }
:dep ndarray = "0.17.2"
:dep num-complex = "0.4.6"
```

## Example


```rust
use rust_rf::{Frequency, FrequencyUnit, SweepType};

let sweep = Frequency::new(
    1.0,
    8.0,
    201,
    FrequencyUnit::GHz,
    SweepType::Linear,
)
.expect("the requested instrument sweep should be valid");

println!(
    "prepared a {}-point sweep; compile live VNA code with --features visa",
    sweep.points(),
);
```
