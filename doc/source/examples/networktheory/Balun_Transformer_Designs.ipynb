# Balun_Transformer_Designs

This is the **rust-rf** edition of the upstream scikit-rf notebook. It demonstrates
assembling and solving a typed microwave circuit graph. Rust examples use typed values and explicit `Result` handling.

The original notebook contained 2 Python code cells. They have been
replaced by a compact, executable Rust workflow instead of retaining stale Python syntax.
The copied data files and figures remain beside the notebook for extending the example.


## Kernel setup

The first cell adds the local crate and the numerical types used by the example.


```rust
:dep rust-rf = { path = "../../../.." }
:dep ndarray = "0.17.2"
:dep num-complex = "0.4.6"
```

## Example


```rust
use num_complex::Complex64;
use rust_rf::circuit::Circuit;
use rust_rf::{Frequency, FrequencyUnit, SweepType};

let frequency = Frequency::new(
    1.0,
    2.0,
    11,
    FrequencyUnit::GHz,
    SweepType::Linear,
)
.expect("the frequency sweep should be valid");
let circuit = Circuit::new(vec![vec![
    Circuit::port(frequency.clone(), "P1", Complex64::new(50.0, 0.0))
        .expect("port P1 should be valid"),
    Circuit::port(frequency.clone(), "P2", Complex64::new(50.0, 0.0))
        .expect("port P2 should be valid"),
    Circuit::port(frequency, "P3", Complex64::new(50.0, 0.0))
        .expect("port P3 should be valid"),
]])
.expect("the three-way intersection should be valid");
let network = circuit.network().expect("the circuit should solve");

println!("{} external ports; S21 = {:.3}", network.ports(), network.s[(0, 1, 0)].re);
```
