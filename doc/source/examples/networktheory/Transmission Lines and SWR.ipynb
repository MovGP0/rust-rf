# Transmission Lines and SWR

This is the **rust-rf** edition of the upstream scikit-rf notebook. It demonstrates
cascading compatible networks with explicit fallible operations. Rust examples use typed values and explicit `Result` handling.

The original notebook contained 2 Python code cells. They have been
replaced by a compact, executable Rust workflow instead of retaining stale Python syntax.
The copied data files and figures remain beside the notebook for extending the example.


## Kernel setup

The first cell adds the local crate and the numerical types used by the example.


```rust
:dep rust-rf = { path = "../../../.." }
:dep ndarray = "0.17.2"
:dep num-complex = "0.4.6"
```

## Example


```rust
use rust_rf::data::DATA;

let line = DATA.line().expect("the embedded line should load");
let cascaded = line.cascade(&line).expect("the two lines should cascade");

println!(
    "cascaded network: {} ports and {} frequency points",
    cascaded.ports(),
    cascaded.frequency_points(),
);
```

## Rust API used

- `Network::cascade`
- `Network::connect`
- `network::cascade_list`
